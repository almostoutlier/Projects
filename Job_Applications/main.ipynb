```python
import matplotlib.pyplot as plt
import sys
import os
import pandas as pd
import re
```

```python
df = pd.read_excel('../Job_Applications/all_emails.xlsx')
df.head()
```

```python
def remove_links(text):
    if pd.isna(text):
        return ""
    return re.sub(r'https?://\S+|www\.\S+', '', text)

df = df.drop(columns=["BCC"])
```

```python
df["Body"] = df["Body"].apply(remove_links)
df = df[~df["From"].str.contains("umass.edu", case=False, na=False)]
df = df[~df["From"].str.contains("Dining", case=False, na=False)]
df = df[~df["From"].str.contains("google.com", case=False, na=False)]
df = df[~df["From"].str.contains("infrastructure", case=False, na=False)]
df = df[~df["From"].str.contains("Doodle", case=False, na=False)]
df = df[~df["From"].str.contains("webex", case=False, na=False)]
df = df[~df['From'].str.contains('instructure.com', case=False, na=False)]
df = df[~df['From'].str.contains('wellfleetinsurance.com', case=False, na=False)]
```

```python
df.head()
```

```python
df.dropna()
```

```python
job_keywords = [
    'application', 'applied', 'job'
]
# Combine into a single regex pattern
pattern = '|'.join(job_keywords)
```

```python
job_df = df[df['Subject'].str.contains(pattern, case=False, na=False) | 
            df['Body'].str.contains(pattern, case=False, na=False)]
```

```python
job_df.head(25)
```

```python
job_df = job_df[~job_df['Subject'].str.contains('Confirm', case=False, na=False)]
job_df = job_df[~job_df['Subject'].str.contains('Verify you', case=False, na=False)]
job_df = job_df[~job_df['Subject'].str.contains('great fit', case=False, na=False)]
job_df = job_df[~job_df['Subject'].str.contains('Welcome', case=False, na=False)]
job_df = job_df[~job_df['Subject'].str.contains('Talent Network', case=False, na=False)]
```

```python
job_df.info()
```

```python
sent_keywords = [
    'application submitted', 'resume submitted',
    'you have applied', 'application received', 'thank you for applying',
    'submission received'
]

# Compile regex pattern
pattern = re.compile('|'.join(sent_keywords), re.IGNORECASE)

# Create new column with flag (True if keywords found)
job_df['Application_Sent'] = job_df.apply(
    lambda row: bool(pattern.search(str(row['Subject']) + " " + str(row['Body']))),
    axis=1
)
```

```python
job_df.head(25)
```

```python
# Define rejection-related keywords
rejection_keywords = [
    'unfortunately', 'not selected', 'we regret', 'declined',
    'rejection', 'unable to move forward', 'not moving forward'
]

# Compile regex pattern
rejection_pattern = re.compile('|'.join(rejection_keywords), re.IGNORECASE)

# Add Rejected column (True if any keyword matched)
job_df['Rejected'] = job_df.apply(
    lambda row: bool(rejection_pattern.search(str(row['Subject']) + " " + str(row['Body']))),
    axis=1
)
```

```python
# Define callback-related keywords
callback_keywords = [
    'schedule a call', 'phone screen', 'video interview', 'follow-up call'
]

# Compile pattern
callback_pattern = re.compile('|'.join(callback_keywords), re.IGNORECASE)

# Add Call_Back column (True if any keyword matches)
job_df['Call_Back'] = job_df.apply(
    lambda row: bool(callback_pattern.search(str(row['Subject']) + " " + str(row['Body']))),
    axis=1
)
```

```python
job_df[job_df['Call_Back']].head(60)
```

```python
application_sent_count = job_df['Application_Sent'].sum()
rejected_count = job_df['Rejected'].sum()
call_back_count = job_df['Call_Back'].sum()

# Print results
print("📊 Email Classification Summary:")
print(f"✅ Applications Sent: {application_sent_count}")
print(f"❌ Rejections:         {rejected_count}")
print(f"📞 Call Backs:         {call_back_count}")
```

```python
total_applied = job_df['Application_Sent'].sum()
total_rejected = job_df['Rejected'].sum()
total_callback = job_df['Call_Back'].sum()

# Ghosted = Applied - (Rejected + Callback)
ghosted = total_applied - (total_rejected + total_callback)


# Print breakdown
print(f"Total Applications: {total_applied}")
print(f"Ghosted: {ghosted}")
print(f"Rejected: {total_rejected}")
print(f"Interviews: {total_callback}")
```

```python
import plotly.graph_objects as go

# Define node labels with counts
labels = [
    "Applications (1399)",        # 0
    "Ghosted (804)",              # 1
    "Rejected (534)",             # 2
    "Interview 1st Stage (61)",   # 3
    "Offers Extended (3)"         # 4
]

# Define flows: source node → target node
sources = [0, 0, 0, 3]  # Applications → Ghosted, Rejected, Interview; Interview → Offer
targets = [1, 2, 3, 4]
values = [804, 534, 61, 3]

# Create Sankey diagram
fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=10,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=labels,
        color="lightblue"
    ),
    link=dict(
        source=sources,
        target=targets,
        value=values,
        color=["#a8c0ff", "#ff8888", "#88cc88", "#ffd700"],  # colors for clarity
        label=[
            f"{labels[s]} → {labels[t]}: {v}" 
            for s, t, v in zip(sources, targets, values)
        ]
    )
)])

fig.update_layout(title_text="📊 Job Application Flow – With Offers", font_size=14)
fig.show()
```
